## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 1: Creating DataFrames

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
A DataFrame is a two-dimensional table with labeled rows and columns. The most common ways to create one in practice are from a dictionary of lists, from a list of dictionaries, or by reading a CSV file with <code>pd.read_csv()</code>.
</div>

### Three common creation patterns

**From a dictionary of lists** — each key becomes a column name and each list becomes the column values. This is the most compact form when you are writing data by hand.

**From a list of dictionaries** — each dictionary is one row. This pattern is common when data arrives as JSON records or when building rows in a loop.

**From a CSV file** — `pd.read_csv("path/to/file.csv")` reads a file from disk and returns a DataFrame. This is the most common starting point in real projects.

Both creation methods should produce identical DataFrames. Count the rows and columns. Identify which columns are text and which are numeric.
</div>

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 3: Selecting Columns

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Column selection narrows the table to only the columns needed for a specific question. Single-bracket notation returns a Series. Double-bracket notation returns a DataFrame.
</div>

### Single column vs multiple columns

```python
df["col"]           # returns a Series (one column)
df[["col1", "col2"]]  # returns a DataFrame (multiple columns)
```

<div style="background:#FFECEC; border-left:6px solid #EB5757; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/caution.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Caution</b><br>
Single brackets return a <b>Series</b>. Double brackets return a <b>DataFrame</b>. Many Pandas methods behave differently depending on which type they receive. If you pass a Series where a DataFrame is expected, you will get unexpected errors. When in doubt, use double brackets.
</div>

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 4: Selecting Rows with loc and iloc

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
<code>.loc[]</code> selects rows and columns by <b>label</b>. <code>.iloc[]</code> selects rows and columns by <b>integer position</b>. Both accept a row selector and an optional column selector separated by a comma.
</div>

### loc vs iloc

| | `.loc[]` | `.iloc[]` |
|---|---|---|
| Uses | Labels / index values | Integer positions (0-based) |
| Row slice end | Inclusive | Exclusive |
| Column selector | Column names | Column positions |

```python
df.loc[0:2, "name":"city"]   # rows 0 to 2 inclusive, columns name to city inclusive
df.iloc[0:3, 1:3]            # rows 0, 1, 2 (position), columns 1 and 2 (position)
df.loc[4]                    # single row by label, returns a Series
df.iloc[0]                   # single row by position, returns a Series
```

Confirm that <code>.loc[0:2]</code> returns rows 0, 1, and 2 (three rows) while <code>.iloc[0:3]</code> also returns three rows. Both slices cover the same data here because the default integer index matches the position.

````python
# loc: label-based
print("--- loc: rows 0 to 2, columns name to city ---")
print(customers.loc[0:2, "name":"city"])

print()

# iloc: position-based
print("--- iloc: rows 0-2 (positions), columns 1-2 (positions) ---")
print(customers.iloc[0:3, 1:3])

print()

# Single row
print("--- single row with loc ---")
print(customers.loc[0])

print()

print("--- single row with iloc ---")
print(customers.iloc[0])
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 5: Filtering Rows

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Filtering keeps only the rows that satisfy a condition. Conditions return a boolean Series, which is then passed back into the DataFrame to select matching rows.
</div>

### Filtering patterns

```python
df[df["col"] > value]                          # single condition
df[(df["col1"] > v1) & (df["col2"] == v2)]     # AND condition
df[(df["col1"] == v1) | (df["col2"] == v2)]    # OR condition
df[df["col"].isin(["A", "B"])]                 # membership test
df[df["col"].between(100, 500)]                # inclusive range
df.query("col > value and col2 == 'text'")     # query string syntax
```

<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>

<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
For each filter, verify that every row in the result satisfies the stated condition. For the range filter, confirm that no row falls outside the bounds.
</div>

<div style="background:#F7FBFF; border-left:6px solid #56CCF2; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/practice.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Try It</b><br>
Add a second condition to the high-spend filter: keep only customers who also have at least 4 visits per month.
</div>

````python
# Single condition
print("--- monthly_spend > 1000 ---")
print(customers[customers["monthly_spend"] > 1000])

print()

# Multiple conditions with &
print("--- monthly_spend > 1000 AND visits_per_month >= 5 ---")
print(customers[(customers["monthly_spend"] > 1000) & (customers["visits_per_month"] >= 5)])

print()

# isin
print("--- city in [Mumbai, Chennai] ---")
print(customers[customers["city"].isin(["Mumbai", "Chennai"])])

print()

# between
print("--- monthly_spend between 900 and 1500 ---")
print(customers[customers["monthly_spend"].between(900, 1500)])

print()

# query
print("--- query: Gold membership with spend > 1200 ---")
print(customers.query("membership == 'Gold' and monthly_spend > 1200"))
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 6: Sorting

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
<code>.sort_values()</code> reorders rows based on one or more columns. <code>.nlargest()</code> and <code>.nsmallest()</code> return the top or bottom n rows by a column without sorting the full DataFrame.
</div>

### Sorting options

```python
df.sort_values("col")                                  # ascending by default
df.sort_values("col", ascending=False)                 # descending
df.sort_values(["col1", "col2"], ascending=[False, True])  # multi-column
df.nlargest(3, "col")                                  # top 3 rows
df.nsmallest(3, "col")                                 # bottom 3 rows
```

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">

Confirm that the top row in the descending sort matches the first row returned by <code>.nlargest(3)</code>.

````python
# Sort by monthly_spend descending
print("--- sort by monthly_spend descending ---")
print(customers.sort_values("monthly_spend", ascending=False))

print()

# Sort by city ascending, then monthly_spend descending
print("--- sort by city asc, monthly_spend desc ---")
print(customers.sort_values(["city", "monthly_spend"], ascending=[True, False]))

print()

# Top 3 spenders
print("--- top 3 spenders ---")
print(customers.nlargest(3, "monthly_spend"))

print()

# Bottom 3 spenders
print("--- bottom 3 spenders ---")
print(customers.nsmallest(3, "monthly_spend"))
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 7: Adding and Removing Columns

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
New columns are created by assigning an expression to a new key on the DataFrame. Columns can be removed with <code>.drop()</code> and renamed with <code>.rename()</code>.
</div>

### Common patterns

```python
df["new_col"] = df["a"] / df["b"]                     # calculated column
df["flag"] = np.where(df["col"] > val, "Yes", "No")   # conditional column
df.drop(columns=["col"])                               # remove a column
df.drop(index=[0, 1])                                  # remove rows by index
df.rename(columns={"old": "new"})                      # rename columns
```

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Add a <code>spend_per_visit</code> column, add a <code>high_spender</code> flag, drop a column, and rename a column.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Manually verify one row: divide that customer's monthly_spend by visits_per_month and confirm it matches the new <code>spend_per_visit</code> value.

````python
# Work on a copy so the base DataFrame stays clean
df = customers.copy()

# Calculated column
df["spend_per_visit"] = (df["monthly_spend"] / df["visits_per_month"]).round(2)

# Conditional column using np.where
df["high_spender"] = np.where(df["monthly_spend"] >= 1500, "Yes", "No")

print("--- after adding columns ---")
print(df[["name", "monthly_spend", "visits_per_month", "spend_per_visit", "high_spender"]])

print()

# Drop a column
df_dropped = df.drop(columns=["high_spender"])
print("--- after dropping high_spender ---")
print(df_dropped.columns.tolist())

print()

# Rename a column
df_renamed = df.rename(columns={"spend_per_visit": "avg_spend_per_visit"})
print("--- after renaming spend_per_visit ---")
print(df_renamed.columns.tolist())
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 8: String Methods

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
The <code>.str</code> accessor exposes string methods that apply to every value in a Series. It is used for cleaning, standardising, and searching text columns.
</div>

### Common str methods

| Method | What it does |
|---|---|
| `.str.lower()` | Convert to lowercase |
| `.str.upper()` | Convert to uppercase |
| `.str.title()` | Capitalise each word |
| `.str.strip()` | Remove leading and trailing whitespace |
| `.str.contains("x")` | Boolean: does value contain x |
| `.str.startswith("x")` | Boolean: does value start with x |
| `.str.replace("a", "b")` | Replace substring |
| `.str.split("x")` | Split into list on delimiter |
| `.str.len()` | Length of each string |

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Introduce messy city values, then clean them using str methods. Use contains and startswith for pattern matching.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
After cleaning, the city column should have consistent capitalisation and no extra whitespace. The contains filter should return only the rows where the pattern matches.
</div>

````python
# Introduce messy city values for demonstration
df_str = customers.copy()
df_str["city"] = ["  mumbai", "DELHI", "Chennai ", "  MUMBAI", "delhi", "CHENNAI", "Mumbai", "  Delhi  "]

print("--- before cleaning ---")
print(df_str["city"].tolist())

# Clean: strip whitespace and apply title case
df_str["city"] = df_str["city"].str.strip().str.title()

print("\n--- after strip and title ---")
print(df_str["city"].tolist())

print()

# contains: find names that contain 'a' (case insensitive)
print("--- names containing 'a' (case insensitive) ---")
print(customers[customers["name"].str.contains("a", case=False)][["name", "city"]])

print()

# startswith
print("--- cities starting with 'M' ---")
print(customers[customers["city"].str.startswith("M")][["name", "city"]])

print()

# str.len
print("--- length of each customer name ---")
print(customers["name"].str.len())
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 9: Handling Missing Values

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Missing values in Pandas are represented as <code>NaN</code>. Detecting, filling, and dropping missing values is one of the most common data cleaning tasks.
</div>

### Missing value toolkit

```python
df.isna()                        # boolean DataFrame: True where NaN
df.notna()                       # boolean DataFrame: True where not NaN
df.isna().sum()                  # count of NaN per column
df.fillna(0)                     # replace NaN with 0
df.fillna(method="ffill")        # forward fill: copy previous value
df.fillna(method="bfill")        # backward fill: copy next value
df.fillna(df.mean(numeric_only=True))  # fill with column mean
df.dropna()                      # drop rows with any NaN
df.dropna(subset=["col"])        # drop rows only if col is NaN
```

````python
# Introduce NaN values
df_nan = customers.copy()
df_nan.loc[1, "monthly_spend"] = np.nan
df_nan.loc[4, "visits_per_month"] = np.nan
df_nan.loc[6, "city"] = np.nan

print("--- NaN count per column ---")
print(df_nan.isna().sum())

print()

# Fill numeric NaN with column mean
mean_spend = df_nan["monthly_spend"].mean()
mean_visits = df_nan["visits_per_month"].mean()
df_filled = df_nan.copy()
df_filled["monthly_spend"] = df_filled["monthly_spend"].fillna(mean_spend).round(2)
df_filled["visits_per_month"] = df_filled["visits_per_month"].fillna(mean_visits).round(1)

print("--- after filling numeric NaN with column mean ---")
print(df_filled[["name", "monthly_spend", "visits_per_month"]].isna().sum())

print()

# Drop rows where city is NaN
df_dropped_nan = df_filled.dropna(subset=["city"])
print("--- after dropna on city column ---")
print(f"Rows before: {len(df_filled)}  Rows after: {len(df_dropped_nan)}")
print(df_dropped_nan[["name", "city"]])
````

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
<code>.groupby()</code> splits the DataFrame into groups by one or more columns and then applies an aggregation function to each group. It answers questions such as "what is the average spend per membership level?"

### GroupBy patterns

```python
df.groupby("col")["target"].mean()             # mean of target per group
df.groupby("col").agg({"target": ["mean", "median", "count"]})  # multiple stats
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 11: Apply and Map

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
<code>.apply()</code> passes each element (or row) through a function. <code>.map()</code> replaces each value in a Series using a dictionary or function. Both allow custom transformations that built-in methods do not cover.
</div>

### apply vs map

| | `.apply()` | `.map()` |
|---|---|---|
| Used on | Series or DataFrame | Series only |
| Input | Function or lambda | Function or dictionary |
| Common use | Custom row or column logic | Value replacement |

```python
series.apply(lambda x: x * 2)              # apply to each element
df.apply(lambda row: row["a"] + row["b"], axis=1)  # apply across columns per row
series.map({"A": 1, "B": 2})               # replace values via dict
```

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Apply a function to categorise spending levels. Map membership tier to a numeric score.

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Verify that each customer's spend label matches their monthly_spend value. Confirm that Gold maps to 3, Silver to 2, and Bronze to 1.
</div>

````python
df_apply = customers.copy()

# apply: categorise spending
def spend_label(spend):
    if spend >= 1500:
        return "High"
    elif spend >= 1000:
        return "Medium"
    else:
        return "Low"

df_apply["spend_category"] = df_apply["monthly_spend"].apply(spend_label)

# Equivalent with a lambda
df_apply["spend_category_lambda"] = df_apply["monthly_spend"].apply(
    lambda x: "High" if x >= 1500 else ("Medium" if x >= 1000 else "Low")
)

print("--- spend category ---")
print(df_apply[["name", "monthly_spend", "spend_category", "spend_category_lambda"]])

print()

# map: membership to numeric level
tier_map = {"Gold": 3, "Silver": 2, "Bronze": 1}
df_apply["membership_level"] = df_apply["membership"].map(tier_map)

print("--- membership numeric level ---")
print(df_apply[["name", "membership", "membership_level"]])

print()

# apply across columns (row-wise with axis=1)
df_apply["spend_per_visit"] = df_apply.apply(
    lambda row: round(row["monthly_spend"] / row["visits_per_month"], 2), axis=1
)
print("--- spend_per_visit via row apply ---")
print(df_apply[["name", "monthly_spend", "visits_per_month", "spend_per_visit"]])
````

## <img src="../../../assets/icons/caution.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Common Mistakes to Avoid

<div style="background:#FFECEC; border-left:6px solid #EB5757; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/caution.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Caution</b><br>
Read through these before the practice task.
</div>

**1. Forgetting double brackets when selecting multiple columns.**
`df["col1", "col2"]` raises a KeyError. Use `df[["col1", "col2"]]`.

**2. Confusing `.loc` (labels) with `.iloc` (positions).**
After filtering and resetting the index, label 0 may not refer to the original first row. When in doubt, check the index with `.index`.

**3. Modifying a view instead of a copy (SettingWithCopyWarning).**
Always use `.copy()` before modifying a filtered subset: `df_subset = df[df["col"] > 0].copy()`.

**4. Using `&` or `|` without parentheses around each condition.**
`df[df["a"] > 1 & df["b"] == 2]` will fail due to operator precedence. Always wrap each condition: `df[(df["a"] > 1) & (df["b"] == 2)]`.

**5. Forgetting that most DataFrame methods return a new object.**
`df.sort_values("col")` does not change `df` in place. You must reassign: `df = df.sort_values("col")` or pass `inplace=True`.

**6. Applying numeric operations to object-dtype columns.**
If a numeric column is stored as `object` dtype (often happens after loading CSV files), operations like `.mean()` will fail. Use `.astype(float)` to convert first.

**7. Confusing `.apply()` on a Series vs a DataFrame.**
On a Series, the function receives one element at a time. On a DataFrame with `axis=1`, the function receives one row as a Series. On a DataFrame with `axis=0`, the function receives one column as a Series. Always confirm which axis you intend.
