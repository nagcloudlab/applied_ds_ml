### Why this matters

- Data work needs values stored with clear names.
- Good variable names make code easier to read.
- Variables can store text, numbers, and other objects.

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 2: Python Lists

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
A list stores multiple values in order.
</div>

### List basics

- Lists use square brackets: `[]`
- The first item is at position `0`
- Lists are useful for storing small collections

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Create a list of monthly purchase amounts.
</div>


````python
purchase_amounts = [1200, 850, 1450, 700, 2200]

print("All purchases:", purchase_amounts)
print("First purchase:", purchase_amounts[0])
print("Number of purchases:", len(purchase_amounts))
print("Total purchase amount:", sum(purchase_amounts))
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 3: Python Dictionaries

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
A dictionary stores key-value pairs.
</div>

### Why dictionaries matter

- They attach labels to values.
- One dictionary can represent one record.
- A row in a table is similar to a labeled record.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Create one customer record as a dictionary.
</div>


````python
customer = {
    "customer_id": 101,
    "name": "Asha",
    "city": "Mumbai",
    "monthly_spend": 1200,
    "visits_per_month": 4,
}

print(customer)
print("Customer city:", customer["city"])
print("Monthly spend:", customer["monthly_spend"])
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 4: NumPy Arrays

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
A NumPy array stores numeric values and supports fast calculations.
</div>

### Why NumPy is useful

- It is efficient for numerical data.
- It supports element-wise operations.
- It is used internally by many data science libraries.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Convert a Python list into a NumPy array.
</div>


````python
import numpy as np

spend_array = np.array([1200, 850, 1450, 700, 2200])

print("Array:", spend_array)
print("Type:", type(spend_array))
print("Average spend:", spend_array.mean())
print("Highest spend:", spend_array.max())
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 5: NumPy Element-Wise Operations

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Element-wise operation means the calculation is applied to every value in the array.
</div>

Example:

- If every customer gets a 10 percent increase in spend estimate, we can multiply all values by `1.10`.
- NumPy does this directly without writing a manual loop.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Apply one calculation to all values in a NumPy array.
</div>


````python
increased_spend = spend_array * 1.10

print("Original spend:", spend_array)
print("After 10 percent increase:", increased_spend.round(1))
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 6: Pandas Series

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
A Pandas Series is one labeled column of data.
</div>

### Simple memory line

- Series = one column
- DataFrame = full table

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Create a Pandas Series for membership types.
</div>


````python
import pandas as pd

membership_series = pd.Series(["Gold", "Silver", "Gold", "Bronze", "Silver"], name="membership")

membership_series
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 7: Pandas DataFrame

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
A DataFrame is a table made of rows and columns.
</div>

### In this example

- One row is one customer.
- Each column stores one detail about the customer.
- This small table is easy to inspect manually.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Create a customer purchase DataFrame.
</div>


````python
customers = pd.DataFrame({
    "customer_id": [101, 102, 103, 104, 105, 106],
    "name": ["Asha", "Ravi", "Meera", "John", "Fatima", "Chen"],
    "city": ["Mumbai", "Delhi", "Chennai", "Mumbai", "Delhi", "Chennai"],
    "membership": ["Gold", "Silver", "Gold", "Bronze", "Silver", "Gold"],
    "monthly_spend": [1200, 850, 1450, 700, 2200, 1600],
    "visits_per_month": [4, 3, 5, 2, 6, 5],
})

customers
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 8: Selecting Columns

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Selecting columns means choosing only the information we need from a table.
</div>

### Why this matters

- Real datasets can have many columns.
- We often inspect only a few columns at a time.
- This keeps analysis focused.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Select customer name, city, and monthly spend.
</div>


````python
selected_columns = customers[["name", "city", "monthly_spend"]]

selected_columns
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 9: Filtering Rows

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Filtering keeps rows that match a condition.
</div>

### Example condition

Keep customers whose monthly spend is greater than `1000`.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Filter high-spending customers.
</div>


````python
high_spend_customers = customers[customers["monthly_spend"] > 1000]

high_spend_customers
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 10: Sorting Data

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Sorting arranges rows based on a column.
</div>

### Why this matters

- Sorting helps find highest and lowest values quickly.
- It is useful before writing observations.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Sort customers by monthly spend from highest to lowest.
</div>


````python
customers_sorted = customers.sort_values("monthly_spend", ascending=False)

customers_sorted
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 11: Adding a New Column

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
A new column can be created from existing columns.
</div>

### Example

If we know monthly spend and visits per month, we can calculate spend per visit.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Create a new column called `spend_per_visit`.
</div>


````python
customers["spend_per_visit"] = customers["monthly_spend"] / customers["visits_per_month"]

customers
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 12: Grouping and Summarizing

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Grouping means splitting data into categories and calculating summaries for each category.
</div>

### Example question

What is the average monthly spend for each membership type?

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Group customers by membership and calculate average monthly spend.
</div>


````python
membership_summary = customers.groupby("membership")["monthly_spend"].mean().reset_index()

membership_summary
````
