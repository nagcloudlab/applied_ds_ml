## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 1: Array Creation

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
NumPy provides several functions for creating arrays. Each serves a different purpose depending on whether you need data from a list, a sequence of numbers, a range, or a filled placeholder.
</div>

### Why this matters

- `np.array()` converts existing data into a NumPy array.
- `np.zeros()` and `np.ones()` create placeholder arrays useful during initialisation.
- `np.full()` fills an array with a specific value.
- `np.arange()` generates evenly spaced integers, similar to Python's `range()`.
- `np.linspace()` generates evenly spaced floats between a start and end value.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Create arrays using each method and print the result.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Notice how <code>arange</code> stops before its end value but <code>linspace</code> includes it. Check the difference between the two approaches for generating a sequence.
</div>

```python
import numpy as np

# From a list of monthly purchase amounts
purchase_amounts = np.array([1200, 850, 1450, 700, 2200, 1600])
print("From list:         ", purchase_amounts)

# Placeholder arrays
zeros = np.zeros(5)
ones = np.ones(5)
filled = np.full(5, 999)
print("Zeros:             ", zeros)
print("Ones:              ", ones)
print("Full (999):        ", filled)

# Sequences
steps = np.arange(0, 10, 2)          # start, stop (exclusive), step
evenly_spaced = np.linspace(0, 1, 6) # start, stop (inclusive), number of points
print("arange(0,10,2):    ", steps)
print("linspace(0,1,6):   ", evenly_spaced)
```

Every NumPy array carries metadata that describes its structure. Checking these properties before operating on an array prevents many common errors.

### Key properties

| Property | What it tells you |
|---|---|

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 3: Indexing and Slicing

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Indexing retrieves a single element. Slicing retrieves a range of elements. Both use square bracket notation and follow 0-based positioning.
</div>

### Rules

- `arr[0]` returns the first element.
- `arr[-1]` returns the last element.
- `arr[1:4]` returns elements at positions 1, 2, and 3 (stop is exclusive).
- `arr[:3]` returns the first three elements.
- `arr[-2:]` returns the last two elements.
- For 2D arrays, use `arr[row, col]` or `arr[row_slice, col_slice]`.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Demonstrate indexing and slicing on 1D and 2D arrays.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Trace each slice back to its position in the original array. Confirm the stop index is always excluded.
</div>

```python
# 1D array: daily customer counts for one week
daily_customers = np.array([120, 95, 140, 88, 175, 200, 160])

print("Full array:          ", daily_customers)
print("First day [0]:       ", daily_customers[0])
print("Last day [-1]:       ", daily_customers[-1])
print("Days 2-4 [1:4]:      ", daily_customers[1:4])
print("First 3 days [:3]:   ", daily_customers[:3])
print("Last 2 days [-2:]:   ", daily_customers[-2:])
print()

# 2D array: 3 stores, 4 weeks of sales
sales = np.array([
    [1200, 1350, 980,  1100],
    [850,  900,  870,  920 ],
    [2100, 2300, 1950, 2200]
])

print("Store 1, Week 2 [0, 1]:      ", sales[0, 1])
print("Store 3, all weeks [2, :]:   ", sales[2, :])
print("All stores, Week 1 [:, 0]:   ", sales[:, 0])
print("Top-left 2x2 block [:2,:2]:")
print(sales[:2, :2])
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 4: Reshaping

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Reshaping changes the structure of an array without changing its data. The total number of elements must stay the same.
</div>

### Key functions

- `.reshape(rows, cols)` returns a new view with a different shape.
- `.flatten()` returns a 1D copy of the array.
- `.ravel()` returns a 1D view where possible (no copy unless needed).
- `.T` transposes the array, swapping rows and columns.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Reshape a 1D array into 2D, flatten it back, and transpose it.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Verify that the element count never changes across all operations. Notice that transposing swaps the shape tuple.
</div>

```python
# 12 months of sales figures
monthly_sales = np.array([1200, 850, 1450, 700, 2200, 1600,
                           1100, 950, 1750, 800, 2000, 1400])

print("Original (1D):")
print(monthly_sales)
print("Shape:", monthly_sales.shape)
print()

# Reshape to 3 quarters x 4 months
quarterly = monthly_sales.reshape(3, 4)
print("Reshaped to (3, 4):")
print(quarterly)
print("Shape:", quarterly.shape)
print()

# Flatten back to 1D
flat = quarterly.flatten()
print("Flattened back to 1D:", flat)
print()

# Transpose
transposed = quarterly.T
print("Transposed (4 months x 3 quarters):")
print(transposed)
print("Shape:", transposed.shape)
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 5: Math and Statistics

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
NumPy arrays have built-in methods for common mathematical and statistical operations. These run on all elements at once without loops.
</div>

### Key functions

- `.sum()` adds all values.
- `.mean()` calculates the average.
- `.std()` calculates the standard deviation.
- `.min()` and `.max()` find the smallest and largest values.
- `.cumsum()` returns the running total.
- `.round(n)` rounds values to `n` decimal places.

### Axis parameter for 2D arrays

- `axis=0` collapses rows (calculates down each column).
- `axis=1` collapses columns (calculates across each row).

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Calculate statistics on a sample array, then show axis behaviour on a 2D array.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
For the 2D example, count the number of results for axis=0 and axis=1. axis=0 should return one value per column and axis=1 should return one value per row.
</div>

```python
spend = np.array([1200, 850, 1450, 700, 2200, 1600])

print("Array:          ", spend)
print("Sum:            ", spend.sum())
print("Mean:           ", spend.mean())
print("Std deviation:  ", spend.std().round(2))
print("Min:            ", spend.min())
print("Max:            ", spend.max())
print("Cumulative sum: ", spend.cumsum())
print("Rounded mean:   ", spend.mean().round(1))
print()

# 2D: 3 stores, 4 weeks
store_sales = np.array([
    [1200, 1350, 980,  1100],
    [850,  900,  870,  920 ],
    [2100, 2300, 1950, 2200]
])

print("2D store sales:")
print(store_sales)
print()
print("Sum per week (axis=0, down columns): ", store_sales.sum(axis=0))
print("Sum per store (axis=1, across rows): ", store_sales.sum(axis=1))
print("Mean per week (axis=0):              ", store_sales.mean(axis=0).round(1))
print("Mean per store (axis=1):             ", store_sales.mean(axis=1).round(1))
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 6: Element-wise Operations

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Element-wise operations apply a calculation to every element in the array simultaneously. No loop is needed.
</div>

### Arithmetic operators

- `arr + value`, `arr - value`, `arr * value`, `arr / value`, `arr ** value`
- When used with two arrays of the same shape, operations are paired position by position.

### Comparison operators

- `arr > value`, `arr < value`, `arr == value`
- These return a boolean array of the same shape.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Demonstrate element-wise arithmetic and comparisons.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Pick one element in the original array and trace its value through each operation. Confirm that the comparison output shows <code>True</code> only for elements that satisfy the condition.
</div>

```python
purchase = np.array([1200, 850, 1450, 700, 2200, 1600])
discount = np.array([  50,  30,   80,  20,  100,   60])

print("Original:              ", purchase)

# Arithmetic with scalar
print("Add 200 (bonus):       ", purchase + 200)
print("Multiply by 1.1 (+10%):", (purchase * 1.1).round(1))
print("Divide by 12 (monthly):", (purchase / 12).round(2))
print("Squared:               ", purchase ** 2)
print()

# Arithmetic with another array (element-wise pairing)
net = purchase - discount
print("Discount array:        ", discount)
print("Net after discount:    ", net)
print()

# Comparisons
print("Greater than 1000:     ", purchase > 1000)
print("Less than 900:         ", purchase < 900)
print("Equal to 1450:         ", purchase == 1450)
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 7: Broadcasting

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Broadcasting is NumPy's rule for automatically expanding a smaller array to match the shape of a larger array during an operation. This avoids the need to manually repeat values.
</div>

### When broadcasting applies

- A scalar is broadcast across all elements (seen in Concept 6).
- A 1D array of length `n` is broadcast across a 2D array with `n` columns.
- Arrays with size `1` in a dimension are stretched to match the other array.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Add a weekly bonus to each store's sales using broadcasting.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Confirm that each column (week) received its corresponding bonus value, not the same value across all columns.
</div>

<div style="background:#FFECEC; border-left:6px solid #EB5757; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/caution.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Caution</b><br>
Broadcasting only works when shapes are compatible. NumPy compares dimensions from the right. If they are not equal and neither is 1, NumPy raises a <code>ValueError: operands could not be broadcast together with shapes ...</code>. Always check <code>.shape</code> before combining two arrays.
</div>

```python
# 3 stores, 4 weeks of sales
store_sales = np.array([
    [1200, 1350, 980,  1100],
    [850,  900,  870,  920 ],
    [2100, 2300, 1950, 2200]
])

# Weekly bonuses (one per week, broadcast across all 3 stores)
weekly_bonus = np.array([50, 75, 100, 60])

print("Store sales shape: ", store_sales.shape)
print("Weekly bonus shape:", weekly_bonus.shape)
print()
print("Original sales:")
print(store_sales)
print()
print("Sales after weekly bonus (broadcasting):")
print(store_sales + weekly_bonus)
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 8: Boolean Indexing and Filtering

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Boolean indexing uses a True/False array (a mask) to select elements from an array. Only elements where the mask is <code>True</code> are returned. This is the NumPy equivalent of filtering rows in a table.
</div>

### Key ideas

- A comparison like `arr > 1000` returns a boolean array of the same shape.
- Passing that boolean array back into `arr[...]` selects matching elements.
- `np.where(condition, value_if_true, value_if_false)` applies conditional logic element-wise.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Filter high-value purchases and use <code>np.where()</code> to label each customer.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Verify that all filtered values exceed the threshold. Check that the <code>np.where()</code> output label matches the condition for each position.
</div>

```python
purchase = np.array([1200, 850, 1450, 700, 2200, 1600])
threshold = 1000

# Create boolean mask
mask = purchase > threshold
print("Purchase amounts:    ", purchase)
print("Boolean mask (>1000):", mask)
print()

# Apply mask to filter
high_value = purchase[mask]
print("High-value purchases:", high_value)
print("Count above threshold:", high_value.size)
print()

# np.where for conditional labeling
labels = np.where(purchase > threshold, "High", "Standard")
print("Customer labels:     ", labels)
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 9: Sorting and Searching

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
NumPy provides functions to sort values, find their sorted positions, retrieve unique values, and locate insertion points in sorted arrays.
</div>

### Key functions

- `np.sort(arr)` returns a sorted copy (ascending by default).
- `np.argsort(arr)` returns the indices that would sort the array.
- `np.unique(arr)` returns sorted unique values.
- `np.searchsorted(sorted_arr, value)` finds where a value would be inserted.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Sort purchase amounts, find the sort order, and identify unique membership tier counts.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Use <code>argsort</code> output to identify which customer (by original position) had the lowest spend. Verify <code>np.unique</code> returns no duplicate values.
</div>

```python
purchase = np.array([1200, 850, 1450, 700, 2200, 1600])
visit_counts = np.array([4, 3, 5, 2, 6, 5, 3, 4, 2, 6, 5, 3])

# Sorting
sorted_asc = np.sort(purchase)
sorted_desc = np.sort(purchase)[::-1]  # reverse the sorted array
print("Original:         ", purchase)
print("Sorted ascending: ", sorted_asc)
print("Sorted descending:", sorted_desc)
print()

# argsort: indices that would sort the array
sort_order = np.argsort(purchase)
print("argsort (indices):", sort_order)
print("Lowest spender is customer at index:", sort_order[0])
print()

# Unique values
unique_visits, counts = np.unique(visit_counts, return_counts=True)
print("Unique visit counts:    ", unique_visits)
print("Frequency of each count:", counts)
print()

# searchsorted
insert_pos = np.searchsorted(sorted_asc, 1300)
print("Insert 1300 at position:", insert_pos, "in", sorted_asc)
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 10: Stacking and Splitting

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Stacking combines arrays. Splitting divides them. These operations are used when building datasets from multiple sources or breaking a dataset into subsets.
</div>

### Key functions

- `np.hstack([a, b])` joins arrays side by side (horizontally).
- `np.vstack([a, b])` stacks arrays on top of each other (vertically).
- `np.concatenate([a, b], axis=...)` general-purpose join along any axis.
- `np.split(arr, n)` divides an array into `n` equal parts.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Combine store data from two regions and then split it back apart.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Verify that <code>vstack</code> increases the number of rows and that splitting recovers the original halves.
</div>

```python
# North region stores: 2 stores, 3 weeks
north = np.array([
    [1200, 1350, 980],
    [850,  900,  870]
])

# South region stores: 2 stores, 3 weeks
south = np.array([
    [2100, 2300, 1950],
    [1600, 1750, 1500]
])

# Vertical stack: all stores, same weeks
all_stores = np.vstack([north, south])
print("vstack (all stores, 3 weeks):")
print(all_stores)
print("Shape:", all_stores.shape)
print()

# Horizontal stack: same stores, more weeks
more_weeks_north = np.array([[1100, 1200], [920, 960]])
north_extended = np.hstack([north, more_weeks_north])
print("hstack (north stores, 5 weeks):")
print(north_extended)
print("Shape:", north_extended.shape)
print()

# Split all_stores back into north and south
regions = np.split(all_stores, 2)
print("After split - North:")
print(regions[0])
print("After split - South:")
print(regions[1])
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 11: Random Number Generation

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
NumPy can generate random data from several distributions. Setting a seed ensures results are reproducible: running the same code again produces the same numbers.
</div>

### Key functions

- `np.random.seed(n)` sets the starting point for the random generator.
- `np.random.randint(low, high, size)` generates random integers.
- `np.random.rand(size)` generates uniform floats between 0 and 1.
- `np.random.randn(size)` generates floats from a standard normal distribution (mean 0, std 1).
- `np.random.choice(arr, size)` samples from an array.
- `np.random.shuffle(arr)` shuffles an array in place.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Generate synthetic customer data with a fixed seed for reproducibility.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Run the cell twice. The output should be identical both times because of the seed. Remove the seed line and run again to see different results.
</div>

```python
np.random.seed(42)

# Random integers: customer visit counts between 1 and 10
visit_counts = np.random.randint(1, 11, size=8)
print("Random visit counts (1-10): ", visit_counts)

# Uniform floats: discount rates between 0 and 1
discount_rates = np.random.rand(6).round(3)
print("Random discount rates (0-1):", discount_rates)

# Normal distribution: spend amounts centred around 1000
spend_normal = (np.random.randn(6) * 200 + 1000).round(0)
print("Normal-dist spend amounts:  ", spend_normal)

# Choice: sample membership tiers
tiers = np.array(["Bronze", "Silver", "Gold"])
sampled_tiers = np.random.choice(tiers, size=6)
print("Sampled membership tiers:   ", sampled_tiers)

# Shuffle a sequence in place
order = np.arange(1, 7)
np.random.shuffle(order)
print("Shuffled order:             ", order)
```

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 12: Useful Utility Functions

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
NumPy provides a set of utility functions that handle common data preparation tasks such as clamping outliers, copying arrays, taking absolute values, and rounding in different directions.
</div>

### Key functions

- `np.clip(arr, min, max)` clamps values to a specified range.
- `np.copy(arr)` creates an independent copy of the array.
- `np.abs(arr)` returns the absolute value of each element.
- `np.ceil(arr)` rounds up to the nearest integer.
- `np.floor(arr)` rounds down to the nearest integer.
- `np.round(arr, n)` rounds to `n` decimal places.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Demonstrate each utility function on a sample of values.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Confirm that <code>clip</code> forces values below the minimum up to the minimum and above the maximum down to the maximum. Check that <code>ceil</code> and <code>floor</code> always round in opposite directions.
</div>

```python
values = np.array([200, 850, 1450, 50, 2200, 1600])
decimal_values = np.array([12.3, -5.7, 8.1, -0.9, 3.5, -14.2])

# Clip: enforce a valid spend range of 100 to 2000
clipped = np.clip(values, 100, 2000)
print("Original:     ", values)
print("Clipped (100-2000):", clipped)
print()

# Copy: independent copy (modifying copy leaves original unchanged)
original = np.array([1, 2, 3])
copied = np.copy(original)
copied[0] = 999
print("Original after modifying copy:", original)
print("Copied (modified):            ", copied)
print()

# Absolute value
print("Decimal values:", decimal_values)
print("Absolute:      ", np.abs(decimal_values))
print()

# Rounding directions
print("np.ceil:  ", np.ceil(decimal_values))
print("np.floor: ", np.floor(decimal_values))
print("np.round: ", np.round(decimal_values, 0))
```

## <img src="../../../assets/icons/caution.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Common Mistakes to Avoid

<div style="background:#FFECEC; border-left:6px solid #EB5757; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/caution.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Caution</b><br>
Read this before starting the practice task.
</div>

- **Modifying a view instead of a copy.** Slicing returns a view by default. Changing the slice changes the original. Use `np.copy()` or `.copy()` when you need an independent array.
- **Forgetting 0-based indexing.** The first element is at index `0`, not `1`. The fifth element is at index `4`.
- **Shape mismatch in operations.** Adding two arrays of incompatible shapes raises a `ValueError`. Always check `.shape` before combining arrays.
- **Assuming `.reshape()` always makes a copy.** `reshape` often returns a view. Modifying the reshaped array may change the original.
- **Integer vs float dtype surprises.** An array created from integers will use integer dtype. Dividing `5 / 2` gives `2` in integer arrays, not `2.5`. Cast with `.astype(float)` when precision matters.
- **Forgetting to set a random seed.** Without a seed, `np.random` results differ every run, making notebooks hard to reproduce and teach from.
