## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 1: What Is Statistics?

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Statistics means using numbers to summarize, understand, and explain data.
</div>

### Practical meaning

If we have many rows of data, we cannot understand everything by reading row after row. Statistics gives us short numerical answers.

### Statistics answers questions like

- What is a typical value?
- How much do values vary?
- What is the smallest or largest value?
- Which value appears most often?
- Do two numeric columns move together?
- Is the data shape balanced or stretched?

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">

Start with a tiny list of spending values so the ideas are easy to see manually.
</div>

````python
small_spend = [700, 850, 1200, 1450, 2200]

print("Small spend list:", small_spend)
print("Number of values:", len(small_spend))
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 3: Mean - The Average

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Mean is the arithmetic average: add all values and divide by the number of values.
</div>

### Formula

```text
mean = sum of values / number of values
```

### When mean is useful

- Values are reasonably balanced
- You want one overall average
- Extreme high or low values are not dominating the data

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Calculate mean manually first, then with Pandas.
</div>

````python
manual_mean = sum(small_spend) / len(small_spend)
pandas_mean = customers["monthly_spend"].mean()

print("Manual mean from small list:", manual_mean)
print("Pandas mean from customer data:", round(pandas_mean, 2))
````

- The manual mean shows the average for the tiny list.
- The Pandas mean shows the average monthly spend in the cleaned customer table.
- Mean is useful, but it can move strongly when one value is very high or very low.

<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Teaching sentence</b><br>

</div>

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 4: Median - The Middle Value

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Median is the middle value after sorting the data.
</div>

### Why median is useful

- It is less affected by very high or very low values.
- It is often better for describing a typical value when data is stretched.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Sort values and calculate median.
</div>

````python
sorted_small_spend = sorted(small_spend)
manual_middle_value = sorted_small_spend[len(sorted_small_spend) // 2]
pandas_median = customers["monthly_spend"].median()

print("Sorted small list:", sorted_small_spend)
print("Manual middle value:", manual_middle_value)
print("Pandas median from customer data:", pandas_median)
````

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Mean and median both describe typical values, but they behave differently when an extreme value is present.
</div>

### Example idea

If one customer spends extremely high, mean will move upward more than median.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Compare mean and median before and after adding one very high value.
</div>

````python
normal_values = [700, 850, 1200, 1450, 2200]
with_extreme_value = [700, 850, 1200, 1450, 8000]

comparison = pd.DataFrame({
    "case": ["normal values", "with extreme high value"],
    "mean": [np.mean(normal_values), np.mean(with_extreme_value)],
    "median": [np.median(normal_values), np.median(with_extreme_value)],
})

comparison
````

## <img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Read the Output: Mean vs Median

- Compare how much the mean changed.
- Compare how much the median changed.
- This is why median is often used when data has unusual high or low values.

<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Mode is the value that appears most often.
</div>

### Mode is useful for

- City
- Membership type

- Any repeated category

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>

<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Spread tells us how different the values are from each other.

### Two groups can have the same average but different spread

Group A values are close together. Group B values are far apart.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Compare two groups with the same mean but different spread.
</div>

````python
group_a = np.array([900, 1000, 1100])
group_b = np.array([200, 1000, 1800])

print("Group A mean:", group_a.mean())
print("Group B mean:", group_b.mean())
print("Group A values:", group_a)
print("Group B values:", group_b)
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 8: Variance - Squared Spread

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Variance measures spread using squared distances from the mean.
</div>

### Why squared distances?

When values are below the mean, their differences are negative. When values are above the mean, their differences are positive.

If we add negative and positive differences directly, they cancel each other.

Squaring solves this:

- Negative differences become positive
- Far values receive more weight
- The result is useful for many statistical calculations

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Show differences from the mean and squared differences step by step.
</div>

````python
values = np.array([8, 10, 12])
mean_value = values.mean()
differences = values - mean_value
squared_differences = differences ** 2

variance_steps = pd.DataFrame({
    "value": values,
    "difference_from_mean": differences,
    "squared_difference": squared_differences,
})

print("Mean:", mean_value)
variance_steps
````

````python
manual_variance = squared_differences.mean()
pandas_variance = customers["monthly_spend"].var()

print("Manual variance for [8, 10, 12]:", manual_variance)
print("Pandas variance for monthly_spend:", round(pandas_variance, 2))
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 9: Standard Deviation - Human-Friendly Spread

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Standard deviation is the square root of variance. It brings spread back to the original unit.
</div>

### Simple explanation

- Variance is useful for math.
- Standard deviation is easier for humans to explain.
- If spend is in rupees, standard deviation is also in rupees.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Calculate standard deviation from variance and compare with Pandas.
</div>

````python
manual_std = np.sqrt(manual_variance)
pandas_std = customers["monthly_spend"].std()

print("Manual standard deviation for [8, 10, 12]:", round(manual_std, 3))
print("Pandas standard deviation for monthly_spend:", round(pandas_std, 2))
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 11: Covariance - Direction of Movement

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Concept</b><br>
Covariance tells us whether two numeric columns tend to move together.
</div>

### Simple interpretation

- Positive covariance: both columns tend to move in the same direction
- Negative covariance: one column tends to go up while the other goes down
- Near zero covariance: no clear same-direction or opposite-direction movement

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Important</b><br>
The covariance number is hard to compare directly because it depends on the units of the columns.
</div>

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Calculate covariance between visits per month and monthly spend.
</div>

````python
covariance_matrix = customers[["visits_per_month", "monthly_spend"]].cov().round(2)

covariance_matrix
````

## <img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Read the Output: Covariance Matrix

The covariance matrix has two types of values:

- Diagonal values compare a column with itself. These are variances.
- Off-diagonal values compare two different columns. These are covariances.

If the covariance between `visits_per_month` and `monthly_spend` is positive, then visits and spend tend to move in the same direction.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">

Calculate correlation for monthly spend and visits per month.

````python
numeric_columns = ["monthly_spend", "visits_per_month"]
correlation_matrix = customers[numeric_columns].corr().round(3)

correlation_matrix
````

## <img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Read the Output: Correlation Matrix

- The diagonal values are `1.000` because each column is perfectly related to itself.
- The off-diagonal value shows the relationship between the two different columns.
- A positive value means the two columns tend to move in the same direction.
- The closer the value is to `1` or `-1`, the stronger the movement.

## <img src="../../../assets/icons/recap.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Statistics Summary Table

- Saying mean and median are always the same.
- Ignoring the effect of unusual high or low values.
- Thinking variance and standard deviation are unrelated.
- Forgetting that variance uses squared distances.
- Treating correlation as proof of cause.
- Looking only at one statistic and not comparing multiple summaries.
- Forgetting to explain statistics in plain language.
