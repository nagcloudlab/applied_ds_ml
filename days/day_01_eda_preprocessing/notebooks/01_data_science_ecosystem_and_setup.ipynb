## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 1: What Is Data Science?

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Simple Definition</b><br>
Data Science uses data to answer questions, find patterns, support decisions, and build data-driven solutions.
</div>

### Why this matters

Without data, people often guess. With data, we can look for evidence, compare options, and explain decisions.

Example:

- Guess: "Gold members are more active."
- Data-based thinking: "Let us compare visits, monthly spend, and membership groups."

### Example questions

- Which membership group has the highest monthly spend?
- Which city has the highest average spending?
- Which customers visit most often?
- Are visits and monthly spend related?

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 2: Data Science, AI, ML, and Deep Learning

These terms are related, but they are not the same.

<div style="background:#EAF3FF; border-left:6px solid #2F80ED; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Key Idea</b><br>
Data science is the full workflow of using data. ML is one part of the larger AI/data world.
</div>

### Simple comparison

- **Data Science**: use data to understand and decide
- **Artificial Intelligence**: create systems that behave intelligently
- **Machine Learning**: systems learn patterns from data
- **Deep Learning**: pattern learning using neural networks

### Easy memory line

```text
Data Science = decision workflow
AI = intelligent behavior
ML = learning from data
Deep Learning = neural-network learning
```

<div style="background:#FFECEC; border-left:6px solid #EB5757; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/caution.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Common Confusion</b><br>
Do not say ML and data science are exactly the same. ML can be used inside data science, but data science also includes cleaning, analysis, visualization, and communication.
</div>

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 6: First NumPy Example

NumPy is commonly used for numerical work.

A NumPy array is like a list of numbers, but it supports fast mathematical operations.

### What we will calculate

- Total purchase amount
- Average purchase amount
- Highest purchase amount

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Create a small array of purchase amounts and calculate simple summary values.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Check whether total, average, and highest value match your manual understanding of the numbers.
</div>

````python
purchase_amounts = np.array([1200, 850, 1450, 700, 2200])

print("Purchase amounts:", purchase_amounts)
print("Total purchase amount:", purchase_amounts.sum())
print("Average purchase amount:", purchase_amounts.mean())
print("Highest purchase amount:", purchase_amounts.max())
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 7: First Pandas Example

Pandas is used to work with table-like data.

A Pandas DataFrame is similar to a spreadsheet table.

### Table language

- Row: one record
- Column: one variable
- DataFrame: full table
- Series: one column

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>

</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Identify how many customers you can see and what information is stored about each customer.

````python
customers = pd.DataFrame({
    "customer_id": [101, 102, 103, 104, 105],
    "city": ["Mumbai", "Delhi", "Chennai", "Mumbai", "Delhi"],
    "membership": ["Gold", "Silver", "Gold", "Bronze", "Silver"],
    "monthly_spend": [1200, 850, 1450, 700, 2200],
    "visits_per_month": [4, 3, 5, 2, 6],
})

customers
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 8: Rows and Columns

In a table:

- One row usually represents one observation.
- One column usually represents one attribute or measurement.

In the customer table:

- One row is one customer.
- Columns describe city, membership, spending, and visits.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">

````python
print("Rows and columns:", customers.shape)
print("Number of rows:", customers.shape[0])
print("Number of columns:", customers.shape[1])
print("Column names:", list(customers.columns))
````

## <img src="../../../assets/icons/concept.svg" width="22" style="vertical-align:middle; margin-right:6px;"> Concept 9: First Simple Visualization

A chart helps us see patterns faster than reading rows manually.

For a first chart, we will compare customer spending.

### Why charts matter

- Tables are good for exact values.
- Charts are good for visual comparison.
- A chart can quickly show the highest, lowest, and unusual values.

<div style="background:#E9F8EF; border-left:6px solid #27AE60; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/code.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Code Lab</b><br>
Create a simple bar chart using Pandas and Matplotlib.
</div>

<div style="background:#FFF8E1; border-left:6px solid #F2C94C; padding:14px; border-radius:6px;">
<img src="../../../assets/icons/read-output.svg" width="22" style="vertical-align:middle; margin-right:6px;"><b>Read the Output</b><br>
Look for the customer with the highest monthly spend and the customer with the lowest monthly spend.
</div>

````python
import matplotlib.pyplot as plt

customers.plot(kind="bar", x="customer_id", y="monthly_spend", legend=False)
plt.title("Monthly Spend by Customer")
plt.xlabel("Customer ID")
plt.ylabel("Monthly Spend")
plt.show()
````
